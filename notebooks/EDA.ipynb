```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("data/yield_climate_final.csv")


```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['yield_kg_ha'], bins=40, kde=True, color='green')
plt.title("Yield Distribution (kg/ha)")
plt.xlabel("Yield (kg/ha)")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("figures/eda/eda_yield_distribution.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(7,6))
corr = data[['t2m','rad','precip','rh2m','yield_kg_ha']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Climate Features and Yield")
plt.tight_layout()
plt.savefig("figures/eda/eda_climate_correlation_heatmap.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y=data['crop'], order=data['crop'].value_counts().index)
plt.title("Sample Count per Crop")
plt.xlabel("Count")
plt.ylabel("Crop")
plt.tight_layout()
plt.savefig("figures/eda/eda_crop_counts.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.pairplot(data[['yield_kg_ha','t2m','rad','precip','rh2m']], 
             kind='reg', plot_kws={'line_kws':{'color':'red'}})
plt.savefig("figures/eda/eda_pairplot_climate_vs_yield.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,12))
top_countries = data['country'].value_counts()

sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Sample Count per Country")
plt.xlabel("Count")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig("figures/eda/eda_country_counts.png", dpi=300)
plt.show()
```
